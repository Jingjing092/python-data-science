# matplotlib库

Matplotlib 是一个用于绘制二维图形的 Python 库，它是一个非常流行的数据可视化工具，特别适用于绘制各种类型的图表、图形和图像。通常和pandas库与NumPy库一起使用。

基本上有两种使用 Matplotlib 的方法：
1. 依靠 pyplot 自动创建和管理图形和轴，并使用 pyplot 函数进行绘图。
2. 显式创建图形和轴，再调用方法。


```python
## 安装包
#!pip install matplotlib -i https://pypi.tuna.tsinghua.edu.cn/simple
## 导入库
import matplotlib.pyplot as plt
## 设置字体为宋体【有一些默认配置不能正常显示中文，则添加以下两行代码】
plt.rcParams['font.family'] = ['serif'] # 设置字体为有衬线字体（宋体是有衬线字体之一）
plt.rcParams['font.serif'] = ['SimSun'] # 设置有衬线字体为宋体
## 下面的是设置字体为黑体
# plt.rcParams['font.family'] = ['sans-serif'] # 设置字体为无衬线字体（黑体是无衬线字体之一）
# plt.rcParams['font.sans-serif'] = ['SimHei'] # 设置无衬线字体为黑体

# 设置公式格式
plt.rcParams['mathtext.fontset'] = 'stix'

# 正常显示负号
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.size']=18 #设置字体字号
plt.rcParams['xtick.labelsize']=16 # 设置横坐标轴字体字号
plt.rcParams['ytick.labelsize']=16 # 设置纵坐标轴字体字号

# 设置刻度朝里
plt.tick_params(which="major",direction='in',length=5,bottom=True,left=True)
```

## plt模块画图
### 基本图形

Matplotlib 中的 pyplot(plt) 模块提供了各种函数，用于绘制不同类型的图表进行可视化。以下是一些常见的 plt 绘图类型：

- 折线图 `plt.plt()` 用于绘制一系列点，通常用于显示数据的趋势。
- 散点图 `plt.scatter()` 用于绘制单个点或一组点，用于观察数据的分布和关系。
- 柱状图 `plt.bar()` 用于绘制垂直柱状图;`plt.barh()` 用于绘制水平柱状图。
- 直方图 `plt.hist()` 用于显示数据分布的频率。
- 饼图 `plt.pie()`用于表示数据的占比。
- 箱线图 `plt.boxplot()` 用于可视化数据的分布、中位数、四分位数等统计信息。

通常，饼图、柱状图用来描述分类数据，其他用于连续变量，散点图可以描述多个变量的关系。

### 例： 作出函数图像$$f(x)=xsin\frac{1}{x}$$

```python
import numpy as np
import math
# 在-0.4和0.4之间划分10000个等分。
x = np.linspace(-0.4,0.4,10000,endpoint=False) # 范围我取的很小是因为在(-0.5,0.5)之外，函数图像就是一个直线，没必要。范围太大你看上去就跟直线似的。可以自己试试。
y = []
for i in x:
    if i==0: 
        y.append(0) # 这里要注意分母不能为0，要不然会报 domain 错误；
        # 这里我直接让y等于0，因为高数都学过极限:趋近于零时，无穷小乘以有界，为0，
    else:
        y.append(i * math.sin(1/i)) 
plt.plot(x,y) # 输入x,y坐标
plt.title("f(x) = xsin(1/x) ") # 设置标题
plt.show() 
```

## 面向对象画图
1. 生成画板 `plt.figure()`

    - 画板大标题 `fig.suptitle("画板标题")`

2. 快速创建图板和一个子图 `fig, ax = plt.subplots()`

    - 在画板里还需要有“图纸”来绘画。也可以理解为坐标系，注意跟下面的axis（坐标轴）要正确区分。

3. 子图上绘画 `ax.plot()`

4. 添加子图 `ax = fig.add_subplot(参数1，参数2，参数3)`

    - 参数1 和参数2是用来对画板划分为几部分；参数3指的是第几部分

    - 例：ax = fig.add_subplot(2,2,1) # 将画板划分为2*2，ax表示的第一个

5. 其他设置

    - 为子图添加标题 `ax.set_title("标题")`

    - 添加图例 `ax.legend()`

    - 在指定位置添加文本 `ax.text(x,y,"str")`

    - 网格是否显示 `ax.grid(True)` # 显示






### 例：python自带泰坦尼克号数据集

本例中是结合pandas库作图，需要对目标列使用 `.values_count()` 函数汇总再画图。

```python
!pip install seaborn
import seaborn as sns
df = sns.load_dataset('titanic')
df.sample(5)
```

对分类数据 survived, pclass, sex 画直方图，并在一张图中显示。

```python
# 创建一个画板
fig = plt.figure(figsize=(12, 4)) # 表示创建的画板的宽度为 12 单位，高度为 4 单位

# 创建三个子图，分别用于 "survived"、"pclass" 和 "sex"
##第一个数字 1 表示总行数。在这个例子中，画板被分为 1 行。
##第二个数字 3 表示总列数。在这个例子中，画板被分为 3 列。
##最后一个数字 表示当前子图在这个网格中的位置。
ax1 = fig.add_subplot(131)  # 第一个子图 ax1 被放置在第一个位置。
ax2 = fig.add_subplot(132)  # 第二个子图
ax3 = fig.add_subplot(133)  # 第三个子图

# 绘制 "survived" 列的柱状图
survived_counts = df['survived'].value_counts()
ax1.bar(survived_counts.index, survived_counts.values)
ax1.set_title('Survived Count')
ax1.set_xticks([0, 1])
ax1.set_xticklabels(['Not Survived', 'Survived'])

# 绘制 "pclass" 列的柱状图
pclass_counts = df['pclass'].value_counts()
ax2.bar(pclass_counts.index, pclass_counts.values)
ax2.set_title('Pclass Count')
ax2.set_xticks([1, 2, 3])
ax2.set_xticklabels(['1st Class', '2nd Class', '3rd Class'])

# 绘制 "sex" 列的柱状图
sex_counts = df['sex'].value_counts()
ax3.bar(sex_counts.index, sex_counts.values)
ax3.set_title('Sex Count')
ax3.legend()

# 调整子图之间的间距
plt.tight_layout()

# 显示图表
plt.show()
```

对 alone 列画饼图，并加上图例。

```python
import matplotlib.pyplot as plt

# 创建一个画板
fig = plt.figure(figsize=(6, 4))

# 创建子图 ax
ax = fig.add_subplot(111)  # 使用 1x1 的子图网格

# 绘制 "alone" 列的饼图，并根据不同类别使用不同颜色
alone_counts = df['alone'].value_counts()
colors = ['aquamarine', 'dodgerblue']
# 绘制饼图
pies, texts, autotexts = ax.pie(alone_counts.values, labels=alone_counts.index, autopct='%1.1f%%', colors=colors)

# 设置子图标题
ax.set_title('Alone or not')

# 添加图例
ax.legend(pies, alone_counts.index)

# 显示图表
plt.show()
```
